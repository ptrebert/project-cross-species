```python
%matplotlib inline

import os as os
import gzip as gz
import collections as col

# What is this?
# Just make sure that the filter_genemodel.py script produces the same output
# as can be readily downloaded from GENCODE for human and mouse
# (modulo minor discrepancies depending on definition of protein coding gene)

genc_dir = '/TL/deep/fhgfs/projects/pebert/thesis/refdata/transcriptome/rawdata'
my_dir = '/TL/deep/fhgfs/projects/pebert/thesis/refdata/transcriptome/fasta'

def read_transcripts(fpath):
    collect = dict()
    tname = ''
    tseq = ''
    print(os.path.basename(fpath))
    with gz.open(fpath, 'rt') as infile:
        for line in infile:
            if not line.strip():
                continue
            if line.startswith('>'):
                if tname:
                    collect[tname.strip()] = tseq.upper()
                tseq = ''
                new_name = line.strip('>').split('|')[0].split('.')[0]
                tname = new_name
            else:
                tseq += line.strip()
    print(len(collect))
    return collect

genc_mmu = read_transcripts(os.path.join(genc_dir, 'gencode.vM1.pc_transcripts.fa.gz'))

my_mmu = read_transcripts(os.path.join(my_dir, 'mmu_mm9_gencode_vM1.transcripts.fa.gz'))

stats = col.Counter()
for k, seq in my_mmu.items():
    try:
        genc_seq = genc_mmu[k]
        if genc_seq == seq:
            stats['match'] += 1
        else:
            stats['no-match'] += 1
    except KeyError:
        stats['no-key'] += 1
print('=== Stats for mouse (Gencode M1)')
print(stats)
        
genc_hsa = read_transcripts(os.path.join(genc_dir, 'gencode.v19.pc_transcripts.fa.gz'))

my_hsa = read_transcripts(os.path.join(my_dir, 'hsa_hg19_gencode_v19.transcripts.fa.gz'))

stats = col.Counter()
for k, seq in my_hsa.items():
    try:
        genc_seq = genc_hsa[k]
        if genc_seq == seq:
            stats['match'] += 1
        else:
            stats['no-match'] += 1
    except KeyError:
        stats['no-key'] += 1
print('=== Stats for human (Gencode 19)')
print(stats)



    
```
